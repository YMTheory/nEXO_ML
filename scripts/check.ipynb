```python
import numpy as np
import matplotlib.pyplot as plt
import plotConfig
```

```python
f_loss_acc = '../training_outputs/loss_acc_preact_resnet.npy'

arr = np.load(f_loss_acc, allow_pickle=True)

y_train_loss = arr[0]
y_train_acc  = arr[1]
y_valid_loss = arr[2]
y_valid_acc  = arr[3]
test_score   = arr[4].tolist()
```
